```python
import MyKmeans
import csv
```

```python
# load dataset
df = spark.read.csv(
    "data/cus-prod-type-totalqty.csv", header=True, mode="DROPMALFORMED"
)

df.show(1)
```

```python
# train df with k = 66
K = 66
kmeans = MyKmeans.MyKmeans(K)
centroids = kmeans.train(df)
result_df = kmeans.predict(df, centroids)
```

```python
result_df.show(5)
```

```python
result_df.groupBy('prediction').sum().sort('prediction').drop('sum(prediction)').show()
```

```python
result = result_df.groupBy('prediction').sum().sort('prediction').drop('sum(prediction)').rdd.map(lambda r: r[:]).collect()
```

```python
result
```

```python
keys = ['k']
keys.extend(df.columns[1:])
result.insert(0, keys)
with open('data/k66-summarize.csv', 'w') as output_file:
    writer = csv.writer(output_file)
    writer.writerows(result)
```
